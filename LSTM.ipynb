```python
# ============================================================
#  LSTM Demand Forecasting — Makerere Food Supply Chain
#  Dataset 1: Cafeteria  → Predict Daily Total Revenue
#  Dataset 2: Kiosk      → Predict Daily Total Revenue
#
#  LSTM is ideal here because:
#  - Data is sequential (700 days)
#  - Demand has weekly & academic-period patterns
#  - LSTM remembers past patterns through its hidden state
# ============================================================

# ============================================================
# STEP 0 — INSTALL (run once in terminal)
# pip install tensorflow pandas numpy matplotlib scikit-learn
# ============================================================

# ============================================================
# STEP 1 — IMPORTS
# ============================================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, Input
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# Fix random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

print("Libraries loaded successfully!")
print(f"TensorFlow version : {tf.__version__}")


# ============================================================
# STEP 2 — HELPER FUNCTIONS
# ============================================================

def clean_currency(series):
    """Remove commas from currency strings and convert to float."""
    return series.astype(str).str.replace(",", "", regex=False).astype(float)


def create_sequences(data, lookback):
    """
    Convert a time series array into (X, y) sequences for LSTM.

    WHAT THIS DOES:
    LSTM does not take raw data — it needs SEQUENCES.
    Given lookback=14, for each day it creates:
        X = the previous 14 days of features
        y = the next day's revenue (what we want to predict)

    Example with lookback=3:
        data = [100, 120, 90, 110, 130]
        X[0] = [100, 120, 90]  →  y[0] = 110
        X[1] = [120, 90, 110]  →  y[1] = 130
    """
    X, y = [], []
    for i in range(lookback, len(data)):
        X.append(data[i - lookback:i])   # past 'lookback' days
        y.append(data[i, 0])             # next day's revenue (column 0)
    return np.array(X), np.array(y)


def evaluate_model(actual, predicted, model_name):
    """Print MAE, RMSE, MAPE for a model."""
    mae  = mean_absolute_error(actual, predicted)
    rmse = np.sqrt(mean_squared_error(actual, predicted))
    mask = actual != 0
    mape = np.mean(np.abs((actual[mask] - predicted[mask]) / actual[mask])) * 100

    print(f"\n{'='*48}")
    print(f"  {model_name} Performance")
    print(f"{'='*48}")
    print(f"  MAE  : {mae:>15,.0f} UGX")
    print(f"  RMSE : {rmse:>15,.0f} UGX")
    print(f"  MAPE : {mape:>14.2f} %")
    return mae, rmse, mape


def build_lstm_model(lookback, n_features, units=64):
    """
    Build a 2-layer LSTM model.

    ARCHITECTURE EXPLAINED:
    Input  → shape (lookback days, n_features)
    LSTM 1 → 64 units, returns full sequence for next LSTM layer
    Dropout→ 20% neurons randomly dropped (prevents overfitting)
    LSTM 2 → 32 units, returns single output vector
    Dropout→ 20% dropout again
    Dense  → 1 unit = the predicted revenue value
    """
    model = Sequential([
        Input(shape=(lookback, n_features)),
        LSTM(units, return_sequences=True),   # Layer 1: learns short patterns
        Dropout(0.2),
        LSTM(units // 2, return_sequences=False), # Layer 2: learns long patterns
        Dropout(0.2),
        Dense(1)                               # Output: single revenue value
    ])

    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
        loss="mae"       # Mean Absolute Error as training loss
    )
    return model


def plot_results(train_history, actual, predicted, title, filename):
    """Plot training loss + actual vs predicted forecast."""
    fig, axes = plt.subplots(1, 2, figsize=(16, 5))
    fig.suptitle(title, fontsize=13, fontweight="bold")

    # Left: Training loss curve
    axes[0].plot(train_history.history["loss"],
                 label="Train Loss", color="#2196F3")
    axes[0].plot(train_history.history["val_loss"],
                 label="Val Loss", color="#FF5722")
    axes[0].set_title("Training Loss Curve")
    axes[0].set_xlabel("Epoch")
    axes[0].set_ylabel("MAE Loss")
    axes[0].legend()
    axes[0].grid(True, alpha=0.3)

    # Right: Actual vs Predicted
    axes[1].plot(actual,    label="Actual",    color="black",   linewidth=2)
    axes[1].plot(predicted, label="Predicted", color="#4CAF50",
                 linewidth=2, linestyle="--")
    axes[1].set_title("Actual vs Predicted (Test Set)")
    axes[1].set_xlabel("Days")
    axes[1].set_ylabel("Daily Revenue (UGX)")
    axes[1].legend()
    axes[1].grid(True, alpha=0.3)
    axes[1].yaxis.set_major_formatter(
        plt.FuncFormatter(lambda x, _: f"{x/1e6:.1f}M"))

    plt.tight_layout()
    plt.savefig(filename, dpi=150, bbox_inches="tight")
    plt.show()
    print(f"Plot saved: {filename}")


# ============================================================
# STEP 3 — LOAD & AGGREGATE CAFETERIA DATASET
# ============================================================
print("\n" + "="*55)
print("  PART A: CAFETERIA LSTM")
print("="*55)

cafe_raw = pd.read_csv("makerere_Cafeteria_synthetic.csv")

# Clean currency columns
for col in ["Revenue_UGX", "Ingredient_Cost_UGX",
            "Waste_Cost_UGX", "Gross_Profit_UGX"]:
    cafe_raw[col] = clean_currency(cafe_raw[col])

cafe_raw["Date"]       = pd.to_datetime(cafe_raw["Date"])
cafe_raw["Is_Weekend"] = cafe_raw["Is_Weekend"].astype(int)

# Encode Academic_Period as number
period_map = {
    "Sem1_Teaching": 1, "Sem1_Exams": 2, "Sem1_Break": 3,
    "Sem2_Teaching": 4, "Sem2_Exams": 5, "Sem2_Break": 6
}
cafe_raw["Academic_Period_num"] = cafe_raw["Academic_Period"].map(period_map).fillna(0)

# Aggregate to DAILY level
# LSTM needs one row per time step (one row per day)
cafe_daily = cafe_raw.groupby("Date").agg(
    Revenue_UGX          = ("Revenue_UGX",           "sum"),
    Portions_Sold        = ("Portions_Sold",          "sum"),
    Portions_Prepared    = ("Portions_Prepared",      "sum"),
    Waste_Portions       = ("Waste_Portions",         "sum"),
    Waste_Pct            = ("Waste_Pct",              "mean"),
    Gross_Profit_UGX     = ("Gross_Profit_UGX",       "sum"),
    Ingredient_Cost_UGX  = ("Ingredient_Cost_UGX",    "sum"),
    Is_Weekend           = ("Is_Weekend",             "first"),
    Academic_Period_num  = ("Academic_Period_num",    "first"),
).reset_index().sort_values("Date")

print(f"Daily cafeteria rows : {len(cafe_daily)}")
print(f"Date range           : {cafe_daily['Date'].min()} → {cafe_daily['Date'].max()}")
print(cafe_daily.head())


# ============================================================
# STEP 4 — PREPARE FEATURES FOR LSTM (CAFETERIA)
# ============================================================
# Features fed into the LSTM at each time step
cafe_features = [
    "Revenue_UGX",         # TARGET is always first (column 0)
    "Portions_Sold",
    "Portions_Prepared",
    "Waste_Portions",
    "Waste_Pct",
    "Ingredient_Cost_UGX",
    "Is_Weekend",
    "Academic_Period_num"
]

cafe_data = cafe_daily[cafe_features].values

# SCALE all values to range [0, 1]
# LSTM trains much better on small numbers
# MinMaxScaler reverses this automatically when we inverse_transform
scaler_cafe = MinMaxScaler(feature_range=(0, 1))
cafe_scaled = scaler_cafe.fit_transform(cafe_data)

print(f"\nFeature matrix shape (scaled): {cafe_scaled.shape}")


# ============================================================
# STEP 5 — CREATE SEQUENCES (CAFETERIA)
# ============================================================
LOOKBACK =  21  # Use past 21 days to predict the next day
                # 14 covers 2 full weeks — captures weekly seasonality

X_cafe, y_cafe = create_sequences(cafe_scaled, LOOKBACK)
print(f"X shape: {X_cafe.shape}  → (samples, lookback, features)")
print(f"y shape: {y_cafe.shape}  → (samples,)")


# ============================================================
# STEP 6 — TRAIN / TEST SPLIT (CAFETERIA)
# Walk-Forward: first 80% train, last 20% test
# ============================================================
split = int(len(X_cafe) * 0.8)

X_train_cafe, X_test_cafe = X_cafe[:split], X_cafe[split:]
y_train_cafe, y_test_cafe = y_cafe[:split], y_cafe[split:]

print(f"\nTrain samples : {len(X_train_cafe)}")
print(f"Test  samples : {len(X_test_cafe)}")


# ============================================================
# STEP 7 — BUILD & TRAIN LSTM (CAFETERIA)
# ============================================================
lstm_cafe = build_lstm_model(
    lookback   = LOOKBACK,
    n_features = len(cafe_features),
    units      = 64
)

lstm_cafe.summary()

# Callbacks — these automatically improve training:
callbacks = [
    EarlyStopping(
        monitor   = "val_loss",
        patience  = 15,        # stop if no improvement for 15 epochs
        restore_best_weights = True
    ),
    ReduceLROnPlateau(
        monitor  = "val_loss",
        factor   = 0.5,        # halve learning rate when stuck
        patience = 7,
        min_lr   = 1e-6
    )
]

print("\nTraining LSTM on Cafeteria data...")
history_cafe = lstm_cafe.fit(
    X_train_cafe, y_train_cafe,
    epochs          = 200,
    batch_size      = 32,
    validation_split= 0.1,    # 10% of train used for validation
    callbacks       = callbacks,
    verbose         = 1
)
print("Cafeteria LSTM training complete!")


# ============================================================
# STEP 8 — PREDICT & INVERSE SCALE (CAFETERIA)
# ============================================================
# Predict on test set
y_pred_cafe_scaled = lstm_cafe.predict(X_test_cafe)

# We need to inverse_transform to get back UGX values
# The scaler expects a full feature matrix, so we pad with zeros
def inverse_scale(scaler, predictions, n_features):
    """Reverse MinMaxScaler on the first column (revenue)."""
    padded = np.zeros((len(predictions), n_features))
    padded[:, 0] = predictions.flatten()
    return scaler.inverse_transform(padded)[:, 0]

y_pred_cafe = inverse_scale(scaler_cafe, y_pred_cafe_scaled, len(cafe_features))
y_true_cafe = inverse_scale(scaler_cafe, y_test_cafe, len(cafe_features))


# ============================================================
# STEP 9 — EVALUATE (CAFETERIA)
# ============================================================
mae_c, rmse_c, mape_c = evaluate_model(
    y_true_cafe, y_pred_cafe, "LSTM — Cafeteria"
)

plot_results(
    history_cafe,
    y_true_cafe,
    y_pred_cafe,
    "LSTM Forecast — Cafeteria Daily Revenue\nMakerere University",
    "lstm_cafeteria_results.png"
)


# ============================================================
# STEP 10 — LOAD & AGGREGATE KIOSK DATASET
# ============================================================
print("\n" + "="*55)
print("  PART B: KIOSK LSTM")
print("="*55)

kiosk_raw = pd.read_csv("makerere_kiosk_synthetic.csv")

kiosk_raw["Total_Amount_UGX"] = clean_currency(kiosk_raw["Total_Amount_UGX"])
kiosk_raw["Unit_Price_UGX"]   = clean_currency(kiosk_raw["Unit_Price_UGX"])
kiosk_raw["Date"]             = pd.to_datetime(kiosk_raw["Date"])
kiosk_raw["Is_Weekend"]       = kiosk_raw["Is_Weekend"].astype(int)
kiosk_raw["Near_Lecture_Start"] = kiosk_raw["Near_Lecture_Start"].astype(int)
kiosk_raw["Academic_Period_num"] = kiosk_raw["Academic_Period"].map(period_map).fillna(0)

# Aggregate to DAILY level
kiosk_daily = kiosk_raw.groupby(["Date","Hour"]).agg(
    Total_Amount_UGX     = ("Total_Amount_UGX",   "sum"),
    Num_Transactions     = ("Transaction_ID",      "count"),
    Avg_Unit_Price       = ("Unit_Price_UGX",      "mean"),
    Total_Quantity       = ("Quantity",            "sum"),
    Is_Weekend           = ("Is_Weekend",          "first"),
    Near_Lecture_Start   = ("Near_Lecture_Start",  "mean"),
    Academic_Period_num  = ("Academic_Period_num", "first"),
).reset_index().sort_values(["Date","Hour"])

print(f"Daily kiosk rows : {len(kiosk_daily)}")
print(kiosk_daily.head())


# ============================================================
# STEP 11 — PREPARE FEATURES FOR LSTM (KIOSK)
# ============================================================
kiosk_features = [
    "Total_Amount_UGX",     # TARGET — always column 0
    "Num_Transactions",
    "Avg_Unit_Price",
    "Total_Quantity",
    "Is_Weekend",
    "Near_Lecture_Start",
    "Academic_Period_num"
]

kiosk_data   = kiosk_daily[kiosk_features].values
scaler_kiosk = MinMaxScaler(feature_range=(0, 1))
kiosk_scaled = scaler_kiosk.fit_transform(kiosk_data)

print(f"Feature matrix shape (scaled): {kiosk_scaled.shape}")


# ============================================================
# STEP 12 — CREATE SEQUENCES (KIOSK)
# ============================================================
X_kiosk, y_kiosk = create_sequences(kiosk_scaled, LOOKBACK)
print(f"X shape: {X_kiosk.shape}")
print(f"y shape: {y_kiosk.shape}")


# ============================================================
# STEP 13 — TRAIN / TEST SPLIT (KIOSK)
# ============================================================
split_k = int(len(X_kiosk) * 0.8)

X_train_kiosk, X_test_kiosk = X_kiosk[:split_k], X_kiosk[split_k:]
y_train_kiosk, y_test_kiosk = y_kiosk[:split_k], y_kiosk[split_k:]

print(f"\nTrain samples : {len(X_train_kiosk)}")
print(f"Test  samples : {len(X_test_kiosk)}")


# ============================================================
# STEP 14 — BUILD & TRAIN LSTM (KIOSK)
# ============================================================
lstm_kiosk = build_lstm_model(
    lookback   = LOOKBACK,
    n_features = len(kiosk_features),
    units      = 64
)

print("\nTraining LSTM on Kiosk data...")
history_kiosk = lstm_kiosk.fit(
    X_train_kiosk, y_train_kiosk,
    epochs           = 100,
    batch_size       = 32,
    validation_split = 0.1,
    callbacks        = callbacks,
    verbose          = 1
)
print("Kiosk LSTM training complete!")


# ============================================================
# STEP 15 — PREDICT & INVERSE SCALE (KIOSK)
# ============================================================
y_pred_kiosk_scaled = lstm_kiosk.predict(X_test_kiosk)

y_pred_kiosk = inverse_scale(scaler_kiosk, y_pred_kiosk_scaled, len(kiosk_features))
y_true_kiosk = inverse_scale(scaler_kiosk, y_test_kiosk,        len(kiosk_features))


# ============================================================
# STEP 16 — EVALUATE (KIOSK)
# ============================================================
mae_k, rmse_k, mape_k = evaluate_model(
    y_true_kiosk, y_pred_kiosk, "LSTM — Kiosk"
)

plot_results(
    history_kiosk,
    y_true_kiosk,
    y_pred_kiosk,
    "LSTM Forecast — Kiosk Daily Revenue\nMakerere University",
    "lstm_kiosk_results.png"
)


# ============================================================
# STEP 17 — FINAL SUMMARY
# ============================================================
print(f"""
╔══════════════════════════════════════════════════════╗
║              LSTM RESULTS SUMMARY                   ║
╠══════════════════════════════════════════════════════╣
║  CAFETERIA (Predicts: Daily Revenue_UGX)            ║
║    MAE  : {mae_c:>15,.0f} UGX               ║
║    RMSE : {rmse_c:>15,.0f} UGX               ║
║    MAPE : {mape_c:>14.2f} %                  ║
╠══════════════════════════════════════════════════════╣
║  KIOSK (Predicts: Daily Total_Amount_UGX)           ║
║    MAE  : {mae_k:>15,.0f} UGX               ║
║    RMSE : {rmse_k:>15,.0f} UGX               ║
║    MAPE : {mape_k:>14.2f} %                  ║
╚══════════════════════════════════════════════════════╝

OUTPUT FILES:
  lstm_cafeteria_results.png  — Loss curve + forecast plot
  lstm_kiosk_results.png      — Loss curve + forecast plot
""")
```

```python
# ── Permutation Feature Importance (no install needed) ──

baseline_mape = mape_c

perm_importance = {}
for i, feat in enumerate(cafe_features):
    X_p = X_test_cafe.copy()
    np.random.shuffle(X_p[:, :, i])
    pred = inverse_scale(scaler_cafe, lstm_cafe.predict(X_p), len(cafe_features))
    mask = y_true_cafe != 0
    mape_perm = np.mean(np.abs((y_true_cafe[mask] - pred[mask]) / y_true_cafe[mask])) * 100
    perm_importance[feat] = round(mape_perm - baseline_mape, 2)

# Plot
import matplotlib.pyplot as plt
sorted_fi = sorted(perm_importance.items(), key=lambda x: x[1], reverse=True)
plt.figure(figsize=(8, 5))
plt.barh([f for f, _ in sorted_fi], [v for _, v in sorted_fi], color="#2196F3")
plt.xlabel("MAPE increase when feature shuffled (%)")
plt.title("Permutation Feature Importance — Cafeteria LSTM")
plt.tight_layout()
plt.savefig("permutation_importance_cafe.png", dpi=150)
plt.show()
print(perm_importance)
```
